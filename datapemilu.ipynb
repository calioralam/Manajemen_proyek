```python
file_path = os.path.join(os.path.expanduser('~'), 'Documents', 'penambangan_data', 'datapemilukpu.csv')
data = pd.read_csv(file_path)
```

```python
# Menampilkan beberapa baris pertama dari dataset
data.head()
```

```python
# Mengelompokkan data berdasarkan partai dan menjumlahkan suara sah caleg
party_votes = data.groupby('NAMA PARTAI POLITIK')['SUARA SAH CALEG'].sum().sort_values(ascending=False)
```

```python
# Membuat plot distribusi suara sah caleg berdasarkan partai
plt.figure(figsize=(10, 6))
sns.barplot(x=party_votes.values, y=party_votes.index, palette='viridis')
plt.title('Distribusi Suara Sah Calon Berdasarkan Partai')
plt.xlabel('Total Suara Sah')
plt.ylabel('Nama Partai')
plt.show()
```

```python
# Menghitung jumlah calon legislatif berdasarkan jenis kelamin
gender_count = data['JENIS KELAMIN'].value_counts()
```

```python
# Membuat pie chart untuk jenis kelamin calon
plt.figure(figsize=(6, 6))
plt.pie(gender_count, labels=gender_count.index, autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff'])
plt.title('Distribusi Calon Berdasarkan Jenis Kelamin')
plt.show()
```

```python
# Memfilter data untuk calon yang terpilih
elected = data[data['TERPILIH ATAU TIDAK'] == 'YA']
```

```python
# Menghitung jumlah caleg terpilih berdasarkan jenis kelamin
gender_elected = elected['JENIS KELAMIN'].value_counts()
```

```python
# Membuat pie chart untuk rasio keterpilihan berdasarkan jenis kelamin
plt.figure(figsize=(6, 6))
plt.pie(gender_elected, labels=gender_elected.index, autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff'])
plt.title('Rasio Keterpilihan Berdasarkan Jenis Kelamin')
plt.show()
```

```python
# Mengelompokkan data berdasarkan daerah pemilihan dan menghitung total suara sah partai dan caleg
district_votes = data.groupby('DAERAH PEMILIHAN')[['SUARA SAH PARTAI', 'SUARA SAH CALEG']].sum()
```

```python
# Membuat plot perbandingan suara sah partai dan caleg per daerah pemilihan
district_votes.plot(kind='bar', figsize=(10, 6))
plt.title('Perbandingan Suara Sah Partai vs Suara Sah Caleg per Daerah Pemilihan')
plt.xlabel('Daerah Pemilihan')
plt.ylabel('Total Suara Sah')
plt.xticks(rotation=0)
plt.show()
```

```python
# Mengelompokkan data berdasarkan partai dan menghitung jumlah kursi yang diperoleh
seat_count = data.groupby('NAMA PARTAI POLITIK')['JUML.PEROLEHAN KURSI'].sum().sort_values(ascending=False)
```

```python
# Membuat plot untuk jumlah kursi per partai
plt.figure(figsize=(10, 6))
sns.barplot(x=seat_count.values, y=seat_count.index, palette='coolwarm')
plt.title('Jumlah Kursi yang Diperoleh Berdasarkan Partai')
plt.xlabel('Jumlah Kursi')
plt.ylabel('Nama Partai')
plt.show()
```
